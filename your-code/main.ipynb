```python
# Your code here:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

base = X.mean().to_frame().T

bmi__vals = np.linspace(X['bmi'].min(), X['bmi'].max(), 200)

x_line = np.repeat(base.values, repeats=200, axis=0)
bmi_idx = list(X.columns).index('bmi')
x_line[:, bmi_idx] = bmi__vals

y_line = model.predict(x_line)

plt.Figure(figsize=(10, 6))
plt.scatter(X['bmi'],y, alpha=0.5)
plt.plot(bmi__vals,y_line)
plt.xlabel('bmi')
plt.ylabel('target')
plt.title('Linear Regression Line')
plt.grid(True, alpha=0.3)
plt.show()
```

Now we will the dataset that we processed and peform linear regression on this data to predict the mpg for each vehicle. Initialize the model in the cell below.

```python
# Your code here:

X = dfauto['horse_power'].values.reshape(-1, 1)
y = dfauto['mpg'].values

plt.figure(figsize=(12, 7))
plt.scatter(X, y, alpha=0.6)
plt.xlabel('horse_power')
plt.ylabel('mpg')
plt.title('Scatter Plot of Horse Power vs MPG')
plt.grid(True, alpha=0.3)
plt.show()

print(f"""Se ve una relación negativa entre la potencia del motor y el rendimiento de combustible (mpg). 
      A medida que la potencia del motor aumenta, el rendimiento de combustible tiende a disminuir. 
      Esto sugiere que los vehículos con motores más potentes suelen ser menos eficientes en términos de consumo de combustible.""")
```

Next, fit the model in the cell below.

```python
# Your code here:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    random_state=seed
)

print(f"Total de datos: {len(X)}")
print(f"Datos de entrenamiento: {len(X_train)} (80%)")
print(f"Datos de prueba: {len(X_test)} (20%)")
```

```python
from sklearn.linear_model import LinearRegression
model_auto = LinearRegression()
model_auto.fit(X_train, y_train)

w0 = model_auto.intercept_
w1 = model_auto.coef_[0]

print(f"Intersección (w0): {w0:.4f}")
print(f"Coeficiente (w1): {w1:.4f}")
print(f'Equacion: mpg = {w0:.4f} + {w1:.4f} * horse_power')
print("\n")
print(f"""El coeficiente w1 es negativo, lo que indica que a medida que la potencia del motor (horse_power) aumenta, el rendimiento de combustible (mpg) disminuye.
Esto es consistente con la observación visual del gráfico de dispersión, donde se ve una relación negativa entre estas dos variables.""")
```

# Challenge 4 - Evaluate the Model

the r squared score of a model tells us how much variation is explained by the model. In a typical dataset, most observations differ from the mean. When we create a model, we are trying to generate an equation that will tell us by how much each observation will differ from the mean. Obviously, the vast majority of models are not perfect. They can only predict some of the variation from the mean but not all of it. We attribute the rest of the difference between the actual value and the mean to random error. We would like random error to explain the as little as possible of the variation. This is why the r squared score is an important metric.

In the next cell, compute the r squared score of the model. Do this by first computing the predicted values and assign them to `y_pred`.

```python
# Your code here:
from sklearn.metrics import mean_squared_error

hw_test = model_auto.predict(X)
mse_test = mean_squared_error(y, hw_test)
r2_test = model_auto.score(X,y)

print(f"El error cuadrático medio es: {mse_test:.4f}\n")
print(f'RMSE (Raiz cuadrada del MSE): {np.sqrt(mse_test):.4f}\n')
print(f"R² del modelo en el conjunto de prueba: {r2_test:.4f}\n")
print(f"""En promedio, las predicciones del modelo se desvían aproximadamente {np.sqrt(mse_test):.2f} unidades de los valores reales.""")
print("\n")
print(f"El modelo explica el {r2_test*100:.2f}% de la variabilidad en el consumo de combustible (mpg) basado en las características del automóvil.")
```

```python
#Visualización de la línea de regresión

plt.figure(figsize = (14, 8))

#datos de entrenamiento y prueba
plt.scatter(X_train,y_train, alpha=0.5, color = 'blue', label='Datos de Entrenamiento')
plt.scatter(X_test, y_test, alpha=0.5, color = 'green', label='Datos de Prueba')

#linea de regresión
X_line = np.linspace(X.min(),X.max(), 200).reshape(-1, 1)
hw_line = model_auto.predict(X_line)
plt.plot(X_line, hw_line, color = 'red', label=f'mpg = {w0:.2f} + {w1:.2f} * horse_power')

plt.xlabel('Caballos de fuerza (horse_power)')
plt.ylabel('Millas por galón (mpg)')
plt.title(f'Regresión Lineal: Horse Power vs MPG (R2= {r2_test:.4f} MSE= {mse_test:.4f} )')
plt.legend()
plt.grid(True, alpha=0.3)  
plt.show()

print("- Puntos azules: datos de entrenamiento")
print("- Puntos verdes: datos de prueba")
print("- Línea roja: modelo de regresión lineal")
```

#### Our next step is to evaluate the model using the test data. We would like to ensure that our model is not overfitting the data. This means that our model will not be able to generalize well outside of the training data.

In the cell below, use the model to generate the predicted values for the training data and assign them to `y_test_pred`. Compute the r squared score for the test data by comparing the oberserved `y_train` data and the predicted `y_test_pred`.

```python
y_test_pred = model_auto.predict(X_test)

# Evaluate the model using R-squared
from sklearn.metrics import r2_score

r2_test = r2_score(y_test, y_test_pred)
r2_test
```
